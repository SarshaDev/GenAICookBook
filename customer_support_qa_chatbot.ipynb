Lets create a prompt template for this chatbot that includes role prompting. Knowledge base information and the user's question

```python
template = """

you are an exceptional customer support chatbot that gently answer questions.

you know the following context information

{chunks_formatted}

Answer the following questions from a customer. use only information from the previous context information. Do not invent stuff

Question : {Query}

Answer:
"""
```

```python
from langchain import PromptTemplate
```

```python
prompt = PromptTemplate(input_variables=["chunks_formatted", "Query"], template=template)
```

#### Retrieval

```python
def retrieve_relevant_context(query):
  docs = db.similarity_search(query)
  chunks = [doc.page_content for doc in docs]
  chunks_formatted = ["\n\n".join(chunks)]
  return chunks_formatted
```

```python
query = "How to check disk usage in Linux"
```

```python
context = retrieve_relevant_context(query)
context
```

```python
prompt_format = prompt.format(chunks_formatted=context, Query=query)

llm = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
```

```python
llm.invoke(prompt_format)
```
